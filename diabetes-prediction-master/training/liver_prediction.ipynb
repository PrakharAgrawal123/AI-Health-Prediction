## Data Visualization

**Import The Library**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Count Plot**

```python
# Target Distribution (To know the result )

sns.countplot(x='Result', data=df)
plt.title("Liver Disease Distribution")
plt.show()
```

```python
# Gender vs Disease (Which one is suffering from disease mostly male of female )

sns.countplot(x='Gender of the patient', hue='Result', data=df)
plt.title("Gender vs Liver Disease")
plt.show()
```

**Histogram**

```python
df.hist(figsize=(12,10), bins=10)
plt.show()
```

**Boxplot (Outliers)**

```python
sns.boxplot(x=df['Total Bilirubin'])
plt.show()

sns.boxplot(x=df['Sgpt Alamine Aminotransferase'])
plt.show()
```

**Correlation Heatmap**

```python
# Encoding 

df['Gender of the patient'] = df['Gender of the patient'].map({'Male':1, 'Female':0})

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

**Box Plot (Feature vs Target)**

```python
sns.barplot(x='Result', y='Total Bilirubin', data=df)
plt.show()
```
